```python
# Read input into lines
with open("input.txt") as fp: lines = fp.read().splitlines()
```

```python
# Main
import re
timestampre = re.compile('Step ([A-Z]) must be finished before step ([A-Z])')

letters = {}
completed = []
todo = []
rom_letters = {}

workers = 2

for line in lines:
    second, first = timestampre.match(line).groups()
    if second not in todo:
        todo.append(second)
    if first not in todo:
        todo.append(first)
        
    if first in letters:
        letters[first].append(second)
    else:
        letters[first] = [second]
out = ""
time = 0

while(len(todo)):
    time += 1
    if time>20:break
    print("Completed", completed)
    print("Letters", letters)
    print("Todo", todo)
    
    for letter in todo:
        if letter not in letters and letter not in rom_letters:
            rom_letters[letter] = ord(letter) - 64
            completed.append(letter)
        else:
            if letter in letters:
                if len(letters[letter]) == 0 and letter not in rom_letters:
                    rom_letters[letter] = ord(letter) - 64
                    completed.append(letter)

    for k in letters.keys():
        if len(letters[k]) == 0:
            if k not in completed:
                completed.append(k)
                
    print("Rom1", rom_letters)
                
    rom = rom_letters.keys()
    rom.sort()
    print("Rom2", rom)
    
    for worker in range(min(workers, len(rom))):
        print('worker doing ', rom[worker])
        rom_letters[rom[worker]] -= 1
        if rom_letters[rom[worker]] == 0:
            out += rom[0]
            todo.remove(rom[0])
            del rom_letters[rom[0]]

        for k in letters.keys():
            if rom[0] in letters[k]:
                letters[k].remove(rom[0])

        
    print("Rom", rom)
    print("Out", out)
    print('time', time) 
    print("")
     
print(out)
print(rom_letters)
print(letters)
```
